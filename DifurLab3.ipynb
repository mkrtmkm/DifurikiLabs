## Задача 1

**Задана лінійна система:**
$$
\begin{cases}
\dot{x} = -2x - y, \\
\dot{y} = 2x + 3y.
\end{cases}
$$

Це автономна система диференціальних рівнянь, яку також можна записати у матричному вигляді $\dot{\mathbf{x}} = A\mathbf{x}$, де матриця системи:
$$A = \begin{pmatrix} -2 & -1 \\ 2 & 3 \end{pmatrix}$$

Диференціальне рівняння фазових траєкторій має вигляд:
$$\frac{dy}{dx} = \frac{\dot{y}}{\dot{x}} = \frac{2x + 3y}{-2x - y}$$

### 1) Кут перетину фазових траєкторій з прямою $y = x$

Щоб знайти кут, під яким траєкторії перетинають пряму $y = x$, підставимо $y = x$ у рівняння фазових траєкторій. Це дасть нам кутовий коефіцієнт дотичної до фазової траєкторії ($k_1$) у точках цієї прямої:
$$k_1 = \frac{dy}{dx}\bigg|_{y=x} = \frac{2x + 3x}{-2x - x} = \frac{5x}{-3x} = -\frac{5}{3}$$

Кутовий коефіцієнт самої прямої $y = x$ дорівнює $k_2 = 1$.
Кут $\alpha$ між двома прямими із заданими кутовими коефіцієнтами шукаємо за формулою:
$$\tan \alpha = \left| \frac{k_1 - k_2}{1 + k_1 k_2} \right|$$

Підставимо наші значення:
$$\tan \alpha = \left| \frac{-\frac{5}{3} - 1}{1 + \left(-\frac{5}{3}\right) \cdot 1} \right| = \left| \frac{-\frac{8}{3}}{1 - \frac{5}{3}} \right| = \left| \frac{-\frac{8}{3}}{-\frac{2}{3}} \right| = 4$$

**Відповідь 1:** Фазові траєкторії перетинають пряму $y = x$ під кутом $\alpha = \arctan(4)$ (що приблизно дорівнює $75.96^\circ$).

### 2) Кут примикання до початку координат $(0,0)$

Щоб дослідити поведінку траєкторій біля початку координат, знайдемо власні значення матриці $A$. Складемо характеристичне рівняння:
$$\det(A - \lambda E) = \begin{vmatrix} -2 - \lambda & -1 \\ 2 & 3 - \lambda \end{vmatrix} = 0$$
$$(-2 - \lambda)(3 - \lambda) - (-1 \cdot 2) = 0$$
$$-6 + 2\lambda - 3\lambda + \lambda^2 + 2 = 0$$
$$\lambda^2 - \lambda - 4 = 0$$

Дискримінант $D = 1 - 4 \cdot 1 \cdot (-4) = 17$.
Власні значення:
$$\lambda_{1,2} = \frac{1 \pm \sqrt{17}}{2}$$
Оскільки $\sqrt{17} \approx 4.12$, маємо:
* $\lambda_1 = \frac{1 + \sqrt{17}}{2} > 0$
* $\lambda_2 = \frac{1 - \sqrt{17}}{2} < 0$

Корені дійсні та мають різні знаки. Отже, точка спокою $(0,0)$ є **сідлом**. У сідла фазові траєкторії примикають до початку координат *лише* вздовж напрямків власних векторів (це так звані сепаратриси).

Знайдемо власні вектори $(A - \lambda E)\mathbf{v} = 0$.
З першого рівняння системи маємо: $(-2 - \lambda)x - y = 0 \Rightarrow y = (-2 - \lambda)x$.
Отже, кутові коефіцієнти прямих, по яких траєкторії входять (або виходять) в $(0,0)$, дорівнюють $k = -2 - \lambda$.

Для $\lambda_1$:
$$k_{\lambda_1} = -2 - \frac{1 + \sqrt{17}}{2} = \frac{-4 - 1 - \sqrt{17}}{2} = \frac{-5 - \sqrt{17}}{2}$$
Для $\lambda_2$:
$$k_{\lambda_2} = -2 - \frac{1 - \sqrt{17}}{2} = \frac{-4 - 1 + \sqrt{17}}{2} = \frac{-5 + \sqrt{17}}{2}$$

**Відповідь 2:** Траєкторії (сепаратриси) примикають до точки $(0,0)$ вздовж двох прямих під кутами:
$\theta_1 = \arctan\left(\frac{-5 - \sqrt{17}}{2}\right)$ та $\theta_2 = \arctan\left(\frac{-5 + \sqrt{17}}{2}\right)$ до осі $Ox$. Інші траєкторії до початку координат не примикають.

```python
import numpy as np
import matplotlib.pyplot as plt

# Налаштування розміру графіка
plt.figure(figsize=(10, 8))

# Створення сітки координат
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)

# Задана система диференціальних рівнянь
U = -2*X - Y
V = 2*X + 3*Y

# Побудова фазового портрета (поля напрямків)
plt.streamplot(X, Y, U, V, color='dodgerblue', density=1.5, linewidth=1, arrowsize=1.5)

# Пряма y = x (з першого завдання)
plt.plot(x, x, 'r--', linewidth=2, label='$y = x$')

# Власні значення та кутові коефіцієнти сепаратрис (з другого завдання)
lam1 = (1 + np.sqrt(17)) / 2
lam2 = (1 - np.sqrt(17)) / 2
k1 = (-5 - np.sqrt(17)) / 2
k2 = (-5 + np.sqrt(17)) / 2

# Побудова сепаратрис (власних векторів)
plt.plot(x, k1 * x, 'g-.', linewidth=2, label=rf'Сепаратриса 1 ($\lambda_1 \approx {lam1:.2f}$)')
plt.plot(x, k2 * x, 'm-.', linewidth=2, label=rf'Сепаратриса 2 ($\lambda_2 \approx {lam2:.2f}$)')

# Налаштування осей та вигляду
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(loc='upper left', fontsize=12)
plt.title("Фазовий портрет системи (Тип: Сідло)", fontsize=16, fontweight='bold')
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

# Показ графіка
plt.show()
```

## Задача 2

**Задана нелінійна автономна система:**
$$
\begin{cases}
\dot{x} = x^2y + xy - 6, \\
\dot{y} = xy + x + y - 5.
\end{cases}
$$

Позначимо праві частини як $P(x,y)$ та $Q(x,y)$:
* $P(x,y) = x^2y + xy - 6$
* $Q(x,y) = xy + x + y - 5$

### 1) Знаходження положень рівноваги та систем першого наближення

Для знаходження положень рівноваги прирівняємо праві частини до нуля:
$$
\begin{cases}
x^2y + xy - 6 = 0, \\
xy + x + y - 5 = 0.
\end{cases}
$$

З другого рівняння виразимо $xy$: $xy = 5 - x - y$. Підставимо це в перше рівняння, попередньо винісши $x$:
$$x(xy) + xy - 6 = 0$$
$$x(5 - x - y) + (5 - x - y) - 6 = 0$$
$$5x - x^2 - xy + 5 - x - y - 6 = 0$$
$$-x^2 + 4x - y(x + 1) - 1 = 0 \implies y(x + 1) = -x^2 + 4x - 1$$

Якщо $x \neq -1$, то $y = \frac{-x^2 + 4x - 1}{x + 1}$.
(Перевірка: якщо $x = -1$, перше рівняння набуває вигляду $y - y - 6 = 0 \implies -6 = 0$, що неможливо).

Підставимо знайдений вираз для $y$ у друге рівняння системи ($xy + x + y - 5 = 0$):
$$x \left( \frac{-x^2 + 4x - 1}{x + 1} \right) + x + \frac{-x^2 + 4x - 1}{x + 1} - 5 = 0$$
Помножимо на $x + 1$:
$$x(-x^2 + 4x - 1) + x(x + 1) + (-x^2 + 4x - 1) - 5(x + 1) = 0$$
$$-x^3 + 4x^2 - x + x^2 + x - x^2 + 4x - 1 - 5x - 5 = 0$$
$$-x^3 + 4x^2 - x - 6 = 0 \implies x^3 - 4x^2 + x + 6 = 0$$

Коренями цього кубічного рівняння є $x_1 = 2$, $x_2 = 3$, $x_3 = -1$. Оскільки $x \neq -1$, залишаються два корені:
* Для $x_1 = 2$: $y_1 = \frac{-4 + 8 - 1}{2 + 1} = \frac{3}{3} = 1$. Точка **$M_1(2, 1)$**.
* Для $x_2 = 3$: $y_2 = \frac{-9 + 12 - 1}{3 + 1} = \frac{2}{4} = 0.5$. Точка **$M_2(3, 0.5)$**.

**Матриця Якобі (для системи першого наближення):**
$$J(x,y) = \begin{pmatrix} \frac{\partial P}{\partial x} & \frac{\partial P}{\partial y} \\ \frac{\partial Q}{\partial x} & \frac{\partial Q}{\partial y} \end{pmatrix} = \begin{pmatrix} 2xy + y & x^2 + x \\ y + 1 & x + 1 \end{pmatrix}$$

**Система першого наближення для $M_1(2, 1)$:**
Обчислимо $J(2,1)$:
$$J(2,1) = \begin{pmatrix} 2(2)(1) + 1 & 2^2 + 2 \\ 1 + 1 & 2 + 1 \end{pmatrix} = \begin{pmatrix} 5 & 6 \\ 2 & 3 \end{pmatrix}$$
Вводячи відхилення $u = x - 2, v = y - 1$, маємо лінеаризовану систему:
$$
\begin{cases}
\dot{u} = 5u + 6v, \\
\dot{v} = 2u + 3v.
\end{cases}
$$

**Система першого наближення для $M_2(3, 0.5)$:**
Обчислимо $J(3,0.5)$:
$$J(3,0.5) = \begin{pmatrix} 2(3)(0.5) + 0.5 & 3^2 + 3 \\ 0.5 + 1 & 3 + 1 \end{pmatrix} = \begin{pmatrix} 3.5 & 12 \\ 1.5 & 4 \end{pmatrix}$$
Вводячи відхилення $u = x - 3, v = y - 0.5$, маємо лінеаризовану систему:
$$
\begin{cases}
\dot{u} = 3.5u + 12v, \\
\dot{v} = 1.5u + 4v.
\end{cases}
$$

### 3) Визначення типу положень рівноваги

**Дослідимо точку $M_1(2, 1)$:**
Характеристичне рівняння для матриці $J(2,1)$:
$$\det \begin{pmatrix} 5 - \lambda & 6 \\ 2 & 3 - \lambda \end{pmatrix} = 0$$
$$(5 - \lambda)(3 - \lambda) - 12 = 0 \implies \lambda^2 - 8\lambda + 3 = 0$$
Дискримінант $D = 64 - 12 = 52$. Корені: $\lambda_{1,2} = \frac{8 \pm \sqrt{52}}{2} = 4 \pm \sqrt{13}$.
Оскільки $\sqrt{13} \approx 3.6$, обидва корені дійсні і додатні ($\lambda_1 > 0, \lambda_2 > 0$).
**Висновок:** Точка $M_1(2, 1)$ є **нестійким вузлом**.

**Дослідимо точку $M_2(3, 0.5)$:**
Характеристичне рівняння для матриці $J(3,0.5)$:
$$\det \begin{pmatrix} 3.5 - \lambda & 12 \\ 1.5 & 4 - \lambda \end{pmatrix} = 0$$
$$(3.5 - \lambda)(4 - \lambda) - 18 = 0 \implies \lambda^2 - 7.5\lambda - 4 = 0$$
Помножимо на 2: $2\lambda^2 - 15\lambda - 8 = 0$.
Дискримінант $D = 225 - 4 \cdot 2 \cdot (-8) = 289 = 17^2$.
Корені: $\lambda_{1,2} = \frac{15 \pm 17}{4}$. Отже, $\lambda_1 = 8 > 0$ і $\lambda_2 = -0.5 < 0$.
Корені дійсні та мають різні знаки.
**Висновок:** Точка $M_2(3, 0.5)$ є **сідлом**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначаємо функції правих частин системи
def P(x, y):
    return x**2 * y + x * y - 6

def Q(x, y):
    return x * y + x + y - 5

# Створюємо фігуру з двома графіками поруч (Subplots)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# ==========================================
# 2) Напрямки векторного поля в точках (m, n)
# ==========================================

# Задаємо цілочисельну сітку від -6 до 6
m_vals = np.arange(-6, 7)
n_vals = np.arange(-6, 7)
M, N = np.meshgrid(m_vals, n_vals)

# Обчислюємо компоненти векторів
U1 = P(M, N)
V1 = Q(M, N)

# Нормалізуємо вектори, щоб вони мали однакову довжину
Norm = np.sqrt(U1**2 + V1**2)
Norm[Norm == 0] = 1 # Уникаємо ділення на нуль
U1_norm = U1 / Norm
V1_norm = V1 / Norm

# Будуємо поле напрямків (quiver plot)
ax1.quiver(M, N, U1_norm, V1_norm, color='teal', alpha=0.8, scale=35)
ax1.set_title(r"Поле напрямків у точках $m, n \in [-6, 6]$", fontsize=14, fontweight='bold')
ax1.set_xlabel("x", fontsize=12)
ax1.set_ylabel("y", fontsize=12)
ax1.set_xticks(m_vals)
ax1.set_yticks(n_vals)
ax1.grid(True, linestyle=':', alpha=0.5)
ax1.axhline(0, color='black', linewidth=1)
ax1.axvline(0, color='black', linewidth=1)


# ==========================================
# 4) Фазові траєкторії в околах положень рівноваги
# ==========================================

# Створюємо більш щільну сітку навколо точок M1(2, 1) та M2(3, 0.5)
x_zoom = np.linspace(0.5, 4.5, 400)
y_zoom = np.linspace(-1, 3, 400)
X_zoom, Y_zoom = np.meshgrid(x_zoom, y_zoom)

U2 = P(X_zoom, Y_zoom)
V2 = Q(X_zoom, Y_zoom)

# Будуємо фазові траєкторії (streamplot)
strm = ax2.streamplot(X_zoom, Y_zoom, U2, V2, color=np.sqrt(U2**2 + V2**2), cmap='viridis', density=1.5, linewidth=1)
fig.colorbar(strm.lines, ax=ax2, label="Швидкість зміни")

# Відмічаємо точки рівноваги
ax2.plot(2, 1, 'ro', markersize=8, label=r'$M_1(2, 1)$ - Нестійкий вузол')
ax2.plot(3, 0.5, 'bo', markersize=8, label=r'$M_2(3, 0.5)$ - Сідло')

ax2.set_title("Фазовий портрет в околі положень рівноваги", fontsize=14, fontweight='bold')
ax2.set_xlabel("x", fontsize=12)
ax2.set_ylabel("y", fontsize=12)
ax2.grid(True, linestyle=':', alpha=0.7)
ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

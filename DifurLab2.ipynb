## Задача 1

**Задане рівняння:**
$$x y' - (2x + 1)y + y^2 = -x^2$$

Це диференціальне рівняння Ріккаті. За умовою задачі його потрібно звести до рівняння Бернуллі шляхом підбору частинного розв'язку.

### Крок 1. Підбір частинного розв'язку $y_1$
Спробуємо знайти частинний розв'язок у вигляді многочлена. Перевіримо $y_1 = x$.
Тоді $y_1' = 1$. Підставимо $y_1$ у вихідне рівняння:
$$x(1) - (2x + 1)x + x^2 = -x^2$$
$$x - 2x^2 - x + x^2 = -x^2$$
$$-x^2 = -x^2$$
Отримана тотожність означає, що **$y_1 = x$** є частинним розв'язком.

### Крок 2. Зведення до рівняння Бернуллі
Використаємо підстановку: $y = y_1 + u(x) = x + u$.
Тоді похідна: $y' = 1 + u'$.
Підставимо це у вихідне рівняння:
$$x(1 + u') - (2x + 1)(x + u) + (x + u)^2 = -x^2$$
Розкриємо дужки:
$$x + x u' - 2x^2 - 2xu - x - u + x^2 + 2xu + u^2 = -x^2$$
Спростимо (скоротимо подібні доданки):
$$x u' - u + u^2 - x^2 = -x^2$$
$$x u' - u + u^2 = 0$$
Поділимо на $x$:
$$u' - \frac{1}{x}u = -\frac{1}{x}u^2$$
Ми отримали **рівняння Бернуллі** з показником $n = 2$.

### Крок 3. Розв'язання рівняння Бернуллі
Для зведення рівняння Бернуллі до лінійного зробимо підстановку:
$$z = u^{1-n} = u^{1-2} = u^{-1} = \frac{1}{u}$$
Звідси $u = \frac{1}{z}$, а $u' = -\frac{z'}{z^2}$.
Підставимо у рівняння Бернуллі:
$$-\frac{z'}{z^2} - \frac{1}{x} \cdot \frac{1}{z} = -\frac{1}{x} \cdot \frac{1}{z^2}$$
Помножимо обидві частини на $-z^2$:
$$z' + \frac{1}{x}z = \frac{1}{x}$$
Це звичайне **лінійне диференціальне рівняння** першого порядку.

### Крок 4. Розв'язання лінійного рівняння
Знайдемо інтегрувальний множник $\mu(x) = e^{\int \frac{1}{x} dx} = e^{\ln|x|} = x$.
Помножимо рівняння на $x$:
$$x z' + z = 1$$
Ліву частину можна згорнути як похідну добутку:
$$(x \cdot z)' = 1$$
Проінтегруємо обидві частини:
$$x \cdot z = x + C$$
$$z = \frac{x + C}{x} = 1 + \frac{C}{x}$$

### Крок 5. Запис загального розв'язку
Повертаємося від $z$ до $u$:
$$u = \frac{1}{z} = \frac{x}{x + C}$$
Повертаємося від $u$ до $y$:
$$y = y_1 + u = x + \frac{x}{x + C}$$

**Відповідь:** Загальний розв'язок рівняння:
$$y(x) = x + \frac{x}{x + C}$$

```python
import sympy as sp
from IPython.display import display, Math

# Визначаємо символьні змінні та функції
x, C = sp.symbols('x C')
y = sp.Function('y')(x)
u = sp.Function('u')(x)
z = sp.Function('z')(x)

print("1. Вихідне рівняння Ріккаті:")
# x*y' - (2x + 1)*y + y^2 = -x^2
eq_riccati = sp.Eq(x * y.diff(x) - (2*x + 1)*y + y**2, -x**2)
display(eq_riccati)

print("\n2. Перевірка частинного розв'язку y1 = x:")
y1 = x
check = eq_riccati.subs({y: y1, y.diff(x): sp.diff(y1, x)}).simplify()
display(Math(r'y_1 = x \Rightarrow \text{Тотожність: }' + str(check)))

print("\n3. Підстановка y = x + u (зведення до Бернуллі):")
y_sub = y1 + u
eq_bernoulli = eq_riccati.subs({y: y_sub, y.diff(x): sp.diff(y_sub, x)}).simplify()
# Переносимо всі члени вліво для зручності
eq_bernoulli = sp.Eq(eq_bernoulli.lhs - eq_bernoulli.rhs, 0).simplify()
display(eq_bernoulli)

print("\n4. Підстановка u = 1/z (зведення до лінійного):")
u_sub = 1/z
# Підставляємо u та u'
eq_linear = eq_bernoulli.subs({u: u_sub, u.diff(x): sp.diff(u_sub, x)})
# Множимо на -z^2 / x щоб привести до канонічного вигляду z' + p(x)z = q(x)
eq_linear_canonical = sp.Eq(sp.simplify(eq_linear.lhs * (-z**2 / x)), 0)
display(eq_linear_canonical)

print("\n5. Розв'язок лінійного рівняння для z(x):")
sol_z = sp.dsolve(eq_linear_canonical, z)
display(sol_z)

print("\n6. Кінцевий загальний розв'язок y(x) = x + 1/z(x):")
# Беремо праву частину розв'язку z(x), знаходимо u = 1/z, і y = x + u
# C1 в sympy замінимо на C для краси
sol_u = 1 / sol_z.rhs
sol_y = y1 + sol_u
sol_y_clean = sol_y.subs('C1', C)
final_solution = sp.Eq(y, sol_y_clean)
display(final_solution)
```

# Задача 2

**Задане рівняння:**
$$y' = -e^x y^2 - y + x^{-\frac{4}{3}} e^{-x}$$

Це диференціальне рівняння Ріккаті загального вигляду $y' = p(x)y^2 + q(x)y + r(x)$, де:
* $p(x) = -e^x$
* $q(x) = -1$
* $r(x) = x^{-\frac{4}{3}} e^{-x}$

За умовою задачі його потрібно спочатку звести до **канонічного вигляду** (тобто $q(x) = 0$).

### Крок 1. Зведення до канонічного вигляду
Щоб позбутися доданка $-y$, використаємо стандартну підстановку:
$$y(x) = u(x) \cdot e^{\int q(x) dx}$$
Оскільки $q(x) = -1$, маємо $\int -1 dx = -x$. Отже, наша підстановка:
$$y = u \cdot e^{-x}$$
Знайдемо похідну:
$$y' = u' e^{-x} - u e^{-x}$$

Підставимо $y$ та $y'$ у вихідне рівняння:
$$u' e^{-x} - u e^{-x} = -e^x (u e^{-x})^2 - (u e^{-x}) + x^{-\frac{4}{3}} e^{-x}$$
$$u' e^{-x} - u e^{-x} = -u^2 e^{-x} - u e^{-x} + x^{-\frac{4}{3}} e^{-x}$$

Скоротимо однакові доданки ($-u e^{-x}$) з обох боків:
$$u' e^{-x} = -u^2 e^{-x} + x^{-\frac{4}{3}} e^{-x}$$

Поділимо обидві частини на $e^{-x}$ (що ніколи не дорівнює нулю):
$$u' = -u^2 + x^{-\frac{4}{3}}$$
або
$$u' + u^2 = x^{-\frac{4}{3}}$$

Ми отримали **канонічне рівняння Ріккаті**.

### Крок 2. Зведення до лінійного рівняння 2-го порядку
Канонічне рівняння Ріккаті вигляду $u' + u^2 = cx^m$ розв'язується за допомогою підстановки:
$$u = \frac{z'}{z}$$
Тоді похідна $u'$ дорівнює:
$$u' = \frac{z'' z - (z')^2}{z^2} = \frac{z''}{z} - \left(\frac{z'}{z}\right)^2 = \frac{z''}{z} - u^2$$

Підставимо це в наше канонічне рівняння:
$$\left(\frac{z''}{z} - u^2\right) + u^2 = x^{-\frac{4}{3}}$$
$$\frac{z''}{z} = x^{-\frac{4}{3}}$$
$$z'' - x^{-\frac{4}{3}}z = 0$$

Це лінійне диференціальне рівняння другого порядку. Знайшовши $z(x)$, ми зможемо повернутися до $u(x)$, а потім і до початкової функції $y(x)$.

```python
import sympy as sp
from IPython.display import display, Math

# Визначаємо символьні змінні та функції
x = sp.symbols('x')
y = sp.Function('y')(x)
u = sp.Function('u')(x)
z = sp.Function('z')(x)

# 1. Оригінальне рівняння
# y' = -e^x * y^2 - y + x^(-4/3) * e^(-x)
eq_riccati = sp.Eq(y.diff(x), -sp.exp(x)*y**2 - y + x**sp.Rational(-4, 3)*sp.exp(-x))
print("1. Вихідне рівняння Ріккаті:")
display(eq_riccati)

# 2. Підстановка y = u * exp(-x) для зведення до канонічного вигляду
y_sub = u * sp.exp(-x)
eq_sub = eq_riccati.subs({y: y_sub, y.diff(x): sp.diff(y_sub, x)})
# Спрощуємо, множачи на exp(x)
eq_canonical = sp.Eq(sp.simplify(eq_sub.lhs * sp.exp(x)), sp.simplify(eq_sub.rhs * sp.exp(x)))
print("\n2. Після підстановки y = u * exp(-x) отримуємо канонічне рівняння:")
display(eq_canonical)

# 3. Підстановка u = z'/z для зведення до лінійного рівняння 2-го порядку
u_sub = z.diff(x) / z
eq_z = eq_canonical.subs({u: u_sub, u.diff(x): sp.diff(u_sub, x)})
# Переносимо все в один бік і множимо на z, щоб отримати z'' - x^(-4/3)z = 0
eq_z_linear = sp.Eq(sp.simplify((eq_z.lhs - eq_z.rhs) * z), 0)
print("\n3. За допомогою підстановки u = z'/z зводимо до лінійного рівняння:")
display(eq_z_linear)

# 4. Розв'язок для z(x)
print("\n4. Розв'язок лінійного рівняння для z(x):")
sol_z = sp.dsolve(eq_z_linear, z)
display(sol_z)

# 5. Повернення до y(x)
print("\n5. Загальний розв'язок вихідного рівняння y(x):")
# Беремо праву частину знайденого z(x)
z_sol_expr = sol_z.rhs

# Знаходимо u = z'/z
u_sol_expr = sp.simplify(z_sol_expr.diff(x) / z_sol_expr)

# Знаходимо y = u * exp(-x)
y_sol_expr = u_sol_expr * sp.exp(-x)

# Виводимо фінальний результат
final_solution = sp.Eq(y, y_sol_expr)
display(final_solution)
```
